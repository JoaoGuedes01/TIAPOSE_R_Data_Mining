## Importar Libraries

```python
import pandas as pd
import requests
import time
```

### Importar o dataset

```python
data = pd.read_csv('../data/TodosDadosTab.csv', index_col=[0])
```

```python
data.head()
```

### Acrescentar a coluna de feriados ao dataframe

```python
# Acrescentar a coluna dos feriados ao dataframe com valor predefinido de 0
data["holiday"] = 0
data.head()
```

### Percorrer todos os dias do df, passa-los pela API "holidays" e guardar o resultado (feriado ou n√£o)

```python
for i in range(len(data)):
    print("row: " + str(i + 1)) # i + 1 porque a length do dataset √© de [0,256] come√ßando em 0
    #Separar string data pelo "-" -> ["2013","04","01"]
    date = data.iloc[i]["date"]
    datesplit = date.split("-")
    year = datesplit[0]
    month = datesplit[1]
    day = datesplit[2]
    print("Data: " + str(date))
    # Pedido GET para a API consoante a linha
    url = "https://holidays.abstractapi.com/v1/?api_key=a05d39b051ce4b2f995da5628acb3806&country=PT&year=" + year + "&month=" + month + "&day=" + day + ""
    
    # Fazer o pedido Get
    x = requests.get(url).json()
    # Se for feriado a resposta vem com 1 objeto
    if len(x)==1:
        print("FERIADO")
        print(x)
        data.iloc[i, data.columns.get_loc('holiday')] = 1
    # Se n√£o for feriado a resposta vem vazia
    else:
        print("NAO FERIADO")
        print(x)
    print("-------------------------------")
    
    # Free tier n√£o suporta mais de 1 request por segundo
    time.sleep(1)
#x = requests.get('https://w3schools.com')
#print(x.status_code)
```

### Todos os dias feriados do dataset

```python
holidays = data.loc[data['holiday'] == 1]
print("Feriados: " + str(len(holidays)))
holidays
```
